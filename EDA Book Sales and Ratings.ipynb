### Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the data

```python
data = pd.read_csv("Books_data_clean.csv")
data
```

```python
data.describe()
```

```python
data = data[data["Publishing Year"] > 1900]
data.describe()
```

```python
data.isna().sum()
```

```python
data.dropna(subset = "Book Name", inplace = True)
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
data.nunique()
```

### Visualization of the data

```python
data.hist("Publishing Year", bins = 50)
plt.xlabel("Publishing Year")
plt.ylabel("Number of Books")
plt.title("Histogram of Publishing Year")
plt.show()
```

```python
data["genre"].value_counts().plot(kind = "bar", cmap = "viridis")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
```

```python
data.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
sns.boxplot(x = "genre", y = "Book_ratings_count", data = data, palette = "viridis")
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book ratings count for Each Genre")
plt.show()
```

```python
plt.scatter(data["sale price"], data["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Units sold")
plt.title("Scatter Plot of Sales Price vs Units Sold")
plt.show()
```

```python
language_counts = data["language_code"].value_counts()
```

```python
plt.pie(language_counts, labels = language_counts.index, startangle = 90, autopct = "%1.1f%%")
plt.show()
```

```python
data.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending = False)
```

```python
data.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending = False)
```

```python
data.groupby("language_code").size().sort_values(ascending = False)
```

```python
data.groupby("Author_Rating")["Book_ratings_count"].max().sort_values(ascending = False)
```

```python
plt.scatter(data["Book_average_rating"], data["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Rating Count")
plt.title("Scatter Plot of Book Average Rating and Book Rating Count")
plt.show()
```

```python
total_gross_sales_by_author = data.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head(10).plot(kind = "bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```

```python
sns.boxplot(x = "Author_Rating", y = "units sold", data = data)
plt.xlabel("Author Rating")
plt.ylabel("Units sold")
plt.title("Box Plot of Units sold per Author Rating")
plt.show()
```

```python
data.groupby("Publishing Year")["units sold"].sum().plot(kind = "line", marker = 'o')
plt.xlabel("Publishing Year")
plt.ylabel("Units Sold")
plt.title("Line Plot of Units sold per Publishing year")
plt.show()
```
